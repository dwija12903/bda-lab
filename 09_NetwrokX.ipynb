```python
# complete graph - all the nodes are connected to each other 
edge_list = [(1,2), (1,3), (1,4), (2,3), (2,4), (3,4)]
G = nx.Graph(edge_list)
nx.draw_spring(G, with_labels=True)
nx.draw_planar(G, with_labels=True)
plt.show()
```

```python
# to calculate shortest path between 2 nodes 
nx.shortest_path(g, 1, 3)
```

```python
nx.has_path(g, 1, 3) 
# check if there is a path between 2 nodes
```

```python
nx.shortest_path_length(g, 1, 3)
# length of the shortest path between 2 nodes
```

```python
nx.shortest_path_length(g, 1, 3, weight='weight')
# shortest path length with weight
```

```python
short = nx.all_pairs_shortest_path(g)
# shortest path between all pairs of nodes

print(list(short))
```

```python
# centality measures - how important a node is in a graph
# degree centrality - number of edges connected to a node
# closeness centrality - how close a node is to all other nodes
# betweenness centrality - how many shortest paths go through a node

nx.degree_centrality(g)
# degree centrality of all nodes
```

```python
nx.closeness_centrality(g)
# closeness centrality of all nodes
```

```python
nx.betweenness_centrality(g)
# betweenness centrality of all nodes
```

```python
# eigenvector centrality - how important a node is based on the importance of its neighbors
nx.eigenvector_centrality(g)
```

```python
dwija_graph = nx.Graph()
dwija_graph.add_edge("Dwija", "Vishal")
dwija_graph.add_edge("Dwija", "Rohan")
dwija_graph.add_edge("Dwija", "Rahul")
dwija_graph.add_node("Krishu")
```

```python
nx.draw_networkx(dwija_graph, with_labels=True)
plt.show()
```

```python
nx.to_dict_of_lists(dwija_graph)
# dictionary of lists
```

```python
# edge list
nx.to_edgelist(dwija_graph)
```

```python
# pandas adajacency matrix
nx.to_pandas_adjacency(dwija_graph)
```

```python
print(list((nx.generate_adjlist(dwija_graph))))
```

```python
nx.clustering(dwija_graph)
# clustering coefficient of all nodes
# clustering coefficient - how many neighbors of a node are connected to each other
```
